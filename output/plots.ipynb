```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("rewardst.txt", sep=" ", header=None, names=["rewards"])
df['reward_roll'] = df['rewards'].rolling(7).mean()
df
```

```python
fig,ax = plt.subplots()

df.plot(ax=ax)
plt.title('Rewards for LavaRoom')
plt.xlabel('episodes')
plt.ylabel('Rewards')
# plt.savefig('../output/lava_reward_small.png')
plt.show()
```
